```python
import pandas as pd
import numpy as np
from google.colab import drive
drive.mount('/content/drive/')

previsores = pd.read_csv('/content/drive/My Drive/CursoIA/base_breast_cancer/entradas_breast.csv')
classe = pd.read_csv('/content/drive/My Drive/CursoIA/base_breast_cancer/saidas_breast.csv')
```

Separando dados para treinamento

```python
from sklearn.model_selection import train_test_split

previsores_treinamento, previsores_teste, classe_treinamento, classe_teste = train_test_split(previsores, classe, test_size=0.25)

```

```python
previsores_treinamento
```

Estruturando rede neural

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
classificador = Sequential()
classificador.add(Dense(units=16, activation='relu',
                        kernel_initializer='random_uniform', input_dim=30))
classificador.add(Dense(units=16, activation='relu',
                        kernel_initializer='random_uniform'))
classificador.add(Dense(units=1, activation='sigmoid'))
```

```python
otimizador = keras.optimizers.Adam(learning_rate = 0.001, decay= 0.0001, clipvalue= 0.5)
classificador.compile(optimizer=otimizador, loss='binary_crossentropy',
                      metrics=['binary_accuracy'])
```

```python
classificador.fit(previsores_treinamento, classe_treinamento,
                  batch_size=10, epochs=100)
```

```python
#Visualizar pesos

pesos0 = classificador.layers[0].get_weights()
print(len(pesos0))
```

```python
previsoes = classificador.predict(previsores_teste)
previsoes = (previsoes > 0.5)

from sklearn.metrics import confusion_matrix, accuracy_score

precisao = accuracy_score(classe_teste, previsoes)
precisao
```

```python
matriz = confusion_matrix(classe_teste, previsoes)
matriz
```

```python
resultado = classificador.evaluate(previsores_teste, classe_teste)
resultado
```

#Rede Neural com K-Fold Classification

```python
# Bibliotecas para esta t√©cnica
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
```

Modularizar cria√ß√£o da rede

```python
def criarRede():
  classificadorKFold = Sequential()
  classificadorKFold.add(Dense(units=16, activation='relu',
                          kernel_initializer='random_uniform', input_dim=30))
  classificadorKFold.add(Dense(units=16, activation='relu',
                          kernel_initializer='random_uniform'))
  classificadorKFold.add(Dense(units=1, activation='sigmoid'))
  otimizadorAdam = keras.optimizers.Adam(learning_rate = 0.001, decay= 0.0001, clipvalue= 0.5)
  classificadorKFold.compile(optimizer=otimizadorAdam, loss='binary_crossentropy',
                        metrics=['binary_accuracy'])
  return classificadorKFold
```

Cria√ß√£o da rede

```python
classificadorFold = KerasClassifier(build_fn = criarRede,
                                    epochs = 100,
                                    batch_size = 10)
resultados = cross_val_score(estimator = classificadorFold,
                             X = previsores,
                             y = classe,
                             cv = 10, scoring = 'accuracy')
```

```python
# Verificando resultado
media = resultados.mean()
print(f'Media: {media}')
desvio = resultados.std()
print(f'Desvio: {desvio}')
```

# Estudando Outfiting e Dropout

```python
from keras.layers  import Dropout

def criarRedeDropout():
  classificadorKFold = Sequential()
  classificadorKFold.add(Dense(units=16, activation='relu',
                          kernel_initializer='random_uniform', input_dim=30))
  classificadorKFold.add(Dropout(0.2))
  classificadorKFold.add(Dense(units=16, activation='relu',
                          kernel_initializer='random_uniform'))
  classificadorKFold.add(Dropout(0.2))
  classificadorKFold.add(Dense(units=1, activation='sigmoid'))
  otimizadorAdam = keras.optimizers.Adam(learning_rate = 0.001, decay= 0.0001, clipvalue= 0.5)
  classificadorKFold.compile(optimizer=otimizadorAdam, loss='binary_crossentropy',
                        metrics=['binary_accuracy'])
  return classificadorKFold
```

Testando t√©cnica implementada

```python
classificadorFold = KerasClassifier(build_fn = criarRede,
                                    epochs = 100,
                                    batch_size = 10)
resultados = cross_val_score(estimator = classificadorFold,
                             X = previsores,
                             y = classe,
                             cv = 10, scoring = 'accuracy')
# Verificando resultado
media = resultados.mean()
print(f'Media: {media}')
desvio = resultados.std()
print(f'Desvio: {desvio}')
```
